```python
import datetime as dt
train=pd.read_csv('/content/drive/MyDrive/archive (2)/fraudTrain.csv')
train['age']=dt.date.today().year-pd.to_datetime(train['dob']).dt.year
train['hour']=pd.to_datetime(train['trans_date_trans_time']).dt.hour
train['day']=pd.to_datetime(train['trans_date_trans_time']).dt.dayofweek
train['month']=pd.to_datetime(train['trans_date_trans_time']).dt.month
train=train[['category','amt','zip','lat','long','city_pop','merch_lat','merch_long','age','hour','day','month','is_fraud']]
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
train['category_encoded'] = label_encoder.fit_transform(train['category'])
train_data =train.drop('category',axis='columns')
```

```python
y_train=train_data['is_fraud'].values
X_train=train_data.drop("is_fraud", axis='columns').values
```

```python
test=pd.read_csv('/content/drive/MyDrive/archive (2)/fraudTest.csv')
test['age']=dt.date.today().year-pd.to_datetime(test['dob']).dt.year
test['hour']=pd.to_datetime(test['trans_date_trans_time']).dt.hour
test['day']=pd.to_datetime(test['trans_date_trans_time']).dt.dayofweek
test['month']=pd.to_datetime(test['trans_date_trans_time']).dt.month
test=test[['category','amt','zip','lat','long','city_pop','merch_lat','merch_long','age','hour','day','month','is_fraud']]
```

```python
from imblearn.over_sampling import SMOTE
from sklearn.ensemble import RandomForestClassifier
method= SMOTE()
X_resampled, y_resampled = method.fit_resample(X_train, y_train)
```

```python
model2 = RandomForestClassifier(random_state=5)
model2.fit(X_resampled,y_resampled)
```

```python
predict_data=[['personal_care',2.86,29209,33.9659,-80.9355,333497,33.986391,-81.200714,34,20,1,1]]
new_df=pd.DataFrame(predict_data,columns=['category','amt','zip','lat','long','city_pop','merch_lat','merch_long','age','hour','day','month'])
category_dummies = pd.get_dummies(new_df['category'], prefix='category')
new_df = new_df.drop('category', axis=1)
new_df = pd.concat([new_df, category_dummies], axis=1)
new_df
```

```python
predictions = model2.predict(new_df)
predictions
```
